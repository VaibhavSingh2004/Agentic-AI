### Creating Tensors

```python
# ------------------------------------------------------------------------------
# empty function allocates only memory do not assign any value in those memory
# locations return already existing value of those locations.
# ------------------------------------------------------------------------------
empty_tensors = torch.empty(2,3) # return 2,3 shape tensors
print(f"Empty tensors: {empty_tensors}, " 
    f"Type of Empty Tensors: {type(empty_tensors)}"
)

print(f"Zero Tensors: {torch.zeros(2,3)}")
print(f"Ones Tensors: {torch.ones(2,3)}")

# ------------------------------------------------------------------------------
# this is used for reproduciblity of same random number on each run withouth this
# random function returns different number on evry run
# ------------------------------------------------------------------------------
torch.manual_seed(100) 
print(f"Random Value Tensors: {torch.rand(2,3)}")

# Convert Python iterables into tensors
print(f"Python Iterable into tensors: {torch.tensor([[1,2,3], [4,5,6]])}")

# Other ways to create tensors
print(f"Using Arange -> {torch.arange(0,10,2)}")
print(f"Using linspace -> {torch.linspace(0,10,10)}")
print(f"Using eye -> {torch.eye(5)}") # identity matrix
print(f"Using full -> {torch.full((3,3), 5)}") # 3X3 matrix and each item value is 5
```

### Tensors Shapes

```python
t = torch.tensor([[1,2,3],[4,5,6]])
print(f"Shape of Tensor t -> {t.shape}")

# How to create same size tensors
print(f"Using empty_like -> {torch.empty_like(t)}")
print(f"Using zeros_like -> {torch.zeros_like(t)}")
print(f"Using ones_like -> {torch.ones_like(t)}")
# print(f"Using rand_like -> {torch.rand_like(t)}")
# NotImplementedError: "check_uniform_bounds" not implemented for 'Long'
# We are getting This error because rand genrates float value between 0 to 1
# but our original tensor t conatains integer value

# Solution:
print(f"Using rand_like -> {torch.rand_like(t, dtype=torch.float16)}")
```

### Tensors DataTypes

```python
tensor = torch.tensor([1.6, 3.4, 8.9])
print(f"Data Typeof tensor -> {tensor.dtype}")

# How to Convert in differnt data_types
print(f"Type Conversion form float to int -> {tensor.to(torch.int16)}")

# How to define datatype while creating tensors
print(f"Tensor of type float32 -> {torch.tensor([1, 2, 4, 5, 6, 7], dtype=torch.float32)}")
```

### Tensor Opeartions

```python
# Opearion between tensor and scalar values
x = torch.tensor([1, 2, 3], dtype=torch.float16)
print(f"Original Tensors -> {x}")
print(f"Addition -> {x + 1}")
print(f"Subtraction -> {x - 1}")
print(f"Multiplication -> {x * 1}")
print(f"Power -> {x ** 2}")
print(f"Mod -> {x % 2}")
print(f"Division -> {x // 2}")
print(f"Equation -> {((x * 2) // 2) + 100}")


# Opearion between two or more tensors
a = torch.rand(2, 3)
b = torch.rand(2, 3)

print(f"Original Tensor A -> {a}")
print(f"Original Tensor B -> {b}")

# Element Wise Operation
print(f"Addition -> {a + b}")
print(f"Subtraction -> {a - b}")
# Simiarly Multiplication, Divison, Modulo etc.

c = torch.tensor([1, -2, 3, -4])
print(f"Absolute -> {torch.abs(c)}")
print(f"Negation -> {torch.neg(c)}")
print(f"Round -> {torch.round(c)}")
print(f"Ceil -> {torch.ceil(c)}")
print(f"Clamp -> {torch.floor(c)}")
print(f"Clamp -> {torch.clamp(c, min=2, max=3)}")
```

```python
# Reduction Operation
e = torch.randint(size=(2, 3), low=0, high=10, dtype=torch.float32)
torch.sum(e)                 # sum
torch.sum(e, dim=0)          # sum along columns
torch.sum(e, dim=1)          # sum along rows
torch.mean(e)                # mean
torch.mean(e, dim=0)         # mean along col
torch.median(e)              # median
torch.max(e)                 # max and min
torch.min(e)
torch.prod(e)                # product
torch.var(e)                 # variance
torch.std(e)                 # standard deviation
torch.argmin(e)              # argmin
torch.argmax(e)              # argmax
```

### Matrix Operation

```python
f = torch.randint(size=(2, 3), low=0, high=10, dtype=torch.float32)
g = torch.randint(size=(3, 2), low=0, high=10, dtype=torch.float32)

torch.matmul(f, g)              # matrix multiplcation
torch.transpose(f, 0, 1)        # transpose

# One Dimensional Vector Needed
vector1 = torch.tensor([1, 2])
vector2 = torch.tensor([3, 4])
torch.dot(vector1, vector2)     # dot product

# Square Matrix needed
h = torch.randint(size=(3, 3), low=0, high=10, dtype=torch.float32)
torch.det(h)                    # determinant
torch.inverse(h)                # inverse

# Special Function
torch.log(f), torch.exp(f), torch.sigmoid(f), torch.relu(f), torch.sqrt(f), torch.softmax(f, dim=0)
# torch.softmax(f, dim=0) Cannot work on long type needed float

# Note: For Inplace Operation (Update the current tensor add `_` in the last of function name e.g., relu_ log_)
torch.log_(f)
```

```python
# How to do Hard Copy of a tensor
f_copy = f.clone()
id(f), id(f_copy) # Both Ids are different
```

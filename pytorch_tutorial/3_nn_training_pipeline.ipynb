```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv(
    "https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv"
)
df.drop(columns=["id", "Unnamed: 32"], inplace=True)
print(df.head()) # top 5 records of dataset
print(df.shape)  # shape of data
```

```python
# Train Test Split 20% data is used for testing rest 80% is used for training
X_train, X_test, y_train, y_test = train_test_split(
    df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2
)
```

```python
# Scale the Training Dataset
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

X_train, y_train
```

```python
# Convert Output Column to integer from Label
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_train, y_test
```

```python
# Convert Numpy Array to Pytorch Tensors
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
print(
    f"Shape of X Train Tensors: {X_train_tensor.shape}\n"
    f"Shape of Y Train Tensors: {y_train_tensor.shape}\n"
    f"Shape of X Test Tensors: {X_test_tensor.shape}\n"
    f"Shape of Y Test Tensors: {y_test_tensor.shape}"
)
```

```python
class MySimpleNN:
    """
    This neural network have just two layers (input and output). No hidden layer.
    """

    def __init__(self, X):

        self.weights = torch.rand(
            X.shape[1], 1, dtype=torch.float64, requires_grad=True
        )
        self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)

    def forward(self, X):
        # Step 1: Linear Combination
        # z = XW + b

        # Diagram:
        #   x1 ----\
        #   x2 -----\
        #   x3 ------> ( weighted sum ) + bias --> z
        #   ... ----/
        #   xn ----/

        print(f"X-Shape: {X.shape} Weights Shape: {self.weights.shape}")
        z = torch.matmul(X, self.weights) + self.bias

        # Step 2: Activation Function (Sigmoid)
        # z --> sigmoid(z) --> y_pred
        # Output range: (0, 1)
        # Interpreted as probability
        y_pred = torch.sigmoid(z)  # activation function
        return y_pred

    def loss_function(self, y_pred, y):

        # Diagram:
        #        y (true label)
        #             |
        #             v
        #   y_pred ---------> Binary Cross Entropy Loss

        # Formula:
        # loss = -[ y*log(y_pred) + (1-y)*log(1-y_pred) ]

        # Clamp predictions to avoid log(0)
        epsilon = 1e-7
        y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

        # Calculate loss - Binary Cross Entropy
        loss = -(
            y_train_tensor * torch.log(y_pred)
            + (1 - y_train_tensor) * torch.log(1 - y_pred)
        ).mean()
        return loss
```

```python
learning_rate = 0.1
epochs = 2500
```

```python
# create model
model = MySimpleNN(X_train_tensor)

# define loop
for epoch in range(epochs):

    # forward pass
    y_pred = model.forward(X_train_tensor)

    # loss calculate
    loss = model.loss_function(y_pred, y_train_tensor)

    # backward pass
    loss.backward()

    # parameters update
    with torch.no_grad():
        model.weights -= learning_rate * model.weights.grad
        model.bias -= learning_rate * model.bias.grad

    # zero gradients
    model.weights.grad.zero_()
    model.bias.grad.zero_()

    # print loss in each epoch
    print(f"Epoch: {epoch + 1}, Loss: {loss.item()}")
```

```python
print(
    f"Model Weight: {model.weights}\n"
    f"Model Bias: {model.bias}"
)
print(f"Shape - Model Weights: {model.weights.shape} Model Bias: {model.bias.shape}")
```

```python
# Evaluation
with torch.no_grad():
    y_pred = model.forward(X_test_tensor)
    y_pred = (y_pred > 0.9).float()
    accuracy = (y_pred == y_test_tensor).float().mean()
    print(f"Accuracy: {accuracy.item()}")
```

### Training Pipeline using nn and optim library

```python
import torch
import torch.nn as nn
import torch.optim as optim # for optimizers
```

```python
df = pd.read_csv(
    "https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv"
)
df.drop(columns=["id", "Unnamed: 32"], inplace=True)
print(df.head())  # top 5 records of dataset
print(df.shape)
```

```python
# Train Test Split 20% data is used for testing rest 80% is used for training
X_train, X_test, y_train, y_test = train_test_split(
    df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2
)

# Scale the Training Dataset
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
print(X_train, y_train)

# Convert Output Column to integer from Label
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)

# Convert Numpy Array to Pytorch Tensors
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)

X_train_tensor = X_train_tensor.float()
y_train_tensor = y_train_tensor.float()
X_test_tensor = X_test_tensor.float()
y_test_tensor = y_test_tensor.float()
```

```python
class MySimpleNN(nn.Module):
    """
    This neural network have just two layers (input and output). No hidden layer.
    """

    def __init__(self, X):
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(X.shape[1], 10), 
            nn.ReLU(),
            nn.Linear(10, 1), 
            nn.Sigmoid()
        )

    def forward(self, X):
        # Step 1: Linear Combination
        # z = XW + b

        # Diagram:
        #   x1 ----\
        #   x2 -----\
        #   x3 ------> ( weighted sum ) + bias --> z
        #   ... ----/
        #   xn ----/

        # Step 2: Activation Function (Sigmoid)
        # z --> sigmoid(z) --> y_pred
        # Output range: (0, 1)
        # Interpreted as probability  # activation function
        return self.network(X)
```

```python
learning_rate = 0.02
epochs = 2500

# create model
model = MySimpleNN(X_train_tensor)

# loss function and optimizer
loss_func = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), learning_rate)

# define loop
for epoch in range(epochs):

    # forward pass
    y_pred = model(X_train_tensor)

    # loss calculate
    loss = loss_func(y_pred, y_train_tensor.view(-1, 1))

    #clear gradients
    optimizer.zero_grad()

    # backward pass
    loss.backward() 

    # parameters update
    optimizer.step()

    # print loss in each epoch
    print(f"Epoch: {epoch + 1}, Loss: {loss.item()}")
```

```python
# Evaluation
with torch.no_grad():
    y_pred = model(X_test_tensor)
    y_pred = (y_pred > 0.9).float()
    accuracy = (y_pred == y_test_tensor).float().mean()
    print(f"Accuracy: {accuracy.item()}")
```
